```python
import pandas as pd
import seaborn as sns
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
df = pd.read_csv("possum.csv")
df.head()
```

```python
x=df[['skullw', 'belly']]
y=df['taill']
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
```

```python
model = LinearRegression()

model.fit(X_train, y_train)
print("skull coefficient:", model.coef_[0])
print("belly coefficient:", model.coef_[1])
print("intercept:", model.intercept_)
print("Prediction=(Skull coefficient)(skull width)+(Belly Coefficient)(Belly girth)+b")
print(f"Prediction = {model.coef_[0]}(Skull width) + {model.coef_[1]}(Belly Girth) + {model.intercept_}")

```
