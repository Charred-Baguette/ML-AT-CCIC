```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
df=pd.read_csv('penguins.csv')
df.dropna
df.head()
```

```python
train_df, test_df = train_test_split(df, test_size=0.4, random_state=42)
train_df = train_df[["culmen_length_mm", "culmen_depth_mm", "flipper_length_mm", "body_mass_g"]]
train_df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
model = KMeans(n_clusters=9)
model.fit(train_df)
cluster_centers = model.cluster_centers_
print(cluster_centers)

x = cluster_centers[:, 0]
y = cluster_centers[:, 1]
sns.scatterplot(x=x, y=y, s=100, color='blue', marker='o')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')

```

```python
x = cluster_centers[:, 1]
y = cluster_centers[:, 2]
sns.scatterplot(x=x, y=y, s=100, color='blue', marker='o')
plt.xlabel('Feature 2')
plt.ylabel('Feature 3')

```

```python
x = cluster_centers[:, 2]
y = cluster_centers[:, 3]
sns.scatterplot(x=x, y=y, s=100, color='blue', marker='o')
plt.xlabel('Feature 3')
plt.ylabel('Feature 4')

```

```python
x = cluster_centers[:, 3]
y = cluster_centers[:, 1]
sns.scatterplot(x=x, y=y, s=100, color='blue', marker='o')
plt.xlabel('Feature 4')
plt.ylabel('Feature 1')
```

```python
tests=[]
X = train_df[["culmen_length_mm", "culmen_depth_mm", "flipper_length_mm", "body_mass_g"]].values
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X)
    tests.append(kmeans.inertia_)
plt.plot(range(1, 11), tests, marker='o')
plt.title('Elbow Graph')
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia')
plt.show()
```

```python
model = KMeans(n_clusters=2)
model.fit(train_df)
cluster_centers = model.cluster_centers_
print(cluster_centers)
labels = model.labels_

sihouette = silhouette_score(X, labels)

print(f"Silhouette Score: {sihouette}")
```
